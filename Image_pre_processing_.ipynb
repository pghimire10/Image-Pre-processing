##Translate Image Randomly


```python
# defining a translation function in this cell
import numpy as np

def translate_image(image, dx, dy):
    """
    This function Translates an image horizontally and/or vertically.

    Args:
        image (PIL.Image.Image): The input image.
        dx (int): The amount of horizontal shift.
        dy (int): The amount of vertical shift.

    Returns:
        PIL.Image.Image: The translated image.
    """
    # Convert the PIL image to a NumPy array
    image_array = np.array(image)

    # Create an empty translated array
    translated_array = np.zeros_like(image_array)
    #assigning image dimention to rows and coloms as rows as height and width i.e dimention of image (h*w)
    rows, cols = image_array.shape[:2]

    """
     logic implemented for translation using two loops in height as rows and width as column,
     (if (i+dx)<rows and (j+dy)<cols) is used to fit original image inside the dimention of
     newly made translated array after translation.
    """
    for i in range(rows):
        for j in range(cols):
            if (i + dx) < rows and (j + dy) < cols:
                translated_array[i + dx][j + dy] = image_array[i][j]

    # Convert the translated array back to a PIL image
    translated_image = Image.fromarray(translated_array)

    return translated_image



# Shift the image horizontally by and vertically by random value
translated_image = translate_image(image, random.randint(0,100),random.randint(0,100))

# Display the translated image
plt.imshow(translated_image)
```

```python
#displaying both images i.e original and translated at the same place

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image)
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(translated_image)
axes[1].set_title('translated_image')
axes[1].axis('off')
plt.show()
```

#Flipping
###Reverse the order of elements in an array along the given axis. The shape of the array is preserved, but the elements are reordered.

##Horizontal Flip

```python
#horizontal flip
def hor_flip_image(image):
  image_array= np.asarray(image)
  hor_flipped_array = np.zeros_like(image_array)
  rows, cols = image_array.shape[:2]

  for i in range(rows):
        for j in range(cols):

                hor_flipped_array[i][-j-1] = image_array[i][j]


  hor_flipped_image = Image.fromarray(hor_flipped_array)

  return hor_flipped_image
hor_flipped_image = hor_flip_image(image)

# Display the translated image
plt.imshow(  hor_flipped_image)



```

```python
# Vertical flip
def ver_flip_image(image):
  image_array= np.asarray(image)
  ver_flipped_array = np.zeros_like(image_array)
  rows, cols = image_array.shape[:2]

  for i in range(rows):
        for j in range(cols):

                ver_flipped_array[-i-1][j] = image_array[i][j]


  ver_flipped_image = Image.fromarray(ver_flipped_array)

  return ver_flipped_image
ver_flipped_image = ver_flip_image(image)
plt.imshow(  ver_flipped_image)
```
